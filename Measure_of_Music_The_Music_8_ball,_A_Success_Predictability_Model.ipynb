```
# Helper function to pull data from Chartmetric API.
def get_chartmetric(endpoint):
  if not endpoint.startswith('https://'):
    endpoint = f'https://api.chartmetric.com/api/{endpoint}'
  
  headers = {
    'Authorization': f'Bearer {chartmetric_creds.get("token")}',
  }
  response = requests.get(endpoint, headers=headers).json()

  return response.get('obj') if 'obj' in response else response
```

# Prep data for analysis

```
# The name of the features being used. This list must match the list inside the
# "pull_features()" function below.
feature_names = [
  'spotify_monthly_listeners',
  'spotify_followers',
  'spotify_playlist_count',
  'instagram_followers',
  'tiktok_followers',
  'tiktok_likes',
  'youtube_subscribers',
  'youtube_views',
  'youtube_monthly_video_views',
  'youtube_playlist_count',
  'cross_platform_performance',
  'audience_growth_vs_engagement',
  'airplay',
  'tv_activity',
]

# Pulls and stores features for the given artist. The features in this function
# must match the "feature_names" list above.
def pull_features(artist):
  features = []
  name = artist['name']
  career_level = artist['career_level']
  chartmetric_id = artist['chartmetric_id']

  # Artist metadata
  artist_stats = get_chartmetric(f'https://api.chartmetric.com/api/artist/{chartmetric_id}')['cm_statistics']

  # Spotify monthly listeners
  features.append(artist_stats['sp_monthly_listeners'] or 0)

  # Spotify followers
  features.append(artist_stats['sp_followers'] or 0)

  # Spotify playlists
  features.append(artist_stats['num_sp_playlists'] or 0)

  # Instagram followers
  features.append(artist_stats['ins_followers'] or 0)

  # Tiktok followers
  features.append(artist_stats['tiktok_followers'] or 0)

  # Tiktok likes
  features.append(artist_stats['tiktok_likes'] or 0)

  # Youtube subscribers
  features.append(artist_stats['ycs_subscribers'] or 0)

  # Youtube views
  features.append(artist_stats['ycs_views'] or 0)

  # Youtube monthly views
  features.append(artist_stats['youtube_monthly_video_views'] or 0)

  # Youtube playlists
  features.append(artist_stats['num_yt_playlists'] or 0)

  # Cross-Platform performance (origial)
  latest_cpp = get_chartmetric(f'https://api.chartmetric.com/api/artist/{chartmetric_id}/cpp?stat=score&latest=true')
  features.append(latest_cpp[0]['score'] if len(latest_cpp) else 0)

  # Audience growth vs Engagement rate (TODO)
  growth = 0
  features.append(growth)

  # Airplay (TODO)
  airplay = 0
  features.append(airplay)

  # Tv activity (TODO)
  # not irrelevant that mid-tier do not tv appearances 
  # tv = get_chartmetric(f'https://api.chartmetric.com/api/artist/{chartmetric_id}/tvmaze')
  # features.append(tv[0]['shows'] if len(tv) else 0)
  features.append(0)

  return features, career_level
```

```
# Features pulled from the APIs for each artist.
raw_features = []

# The career level of each artist.
career_levels = []

artists_for_analysis = artists # or "sample_artists"
for artist in artists_for_analysis:
  features, career_level = pull_features(artist)
  raw_features.append(features)
  career_levels.append(career_level)

print('Feature names:', feature_names)
print('Career levels:', career_levels)
print('Raw features:', raw_features)
```

```
# Create a Pandas dataframe from raw features.
dataset = pd.DataFrame(raw_features, columns=feature_names)

# Peak into the dataset to see what it looks like.
dataset.head()
```
