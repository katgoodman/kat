#Data Cleaning

```python
#1: First, we want to import the libraries we will need.
#2: Then we want to add our raw github and label it as a DataFrame
# We are using John Hopkins data on deaths
#This is the method for importing a url into the Google Colab as a dataframe

import pandas as pd
import io
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
john_hopkins_covid_cases= pd.read_csv(url)
```

#Data Cleaning

```python
#5: We will inspect the John Hopkins Dataframe
pd.set_option('display.max_columns',999)
john_hopkins_covid_cases.head()
# we will remove all date columns before march 1st
```

```python

#6: Let's drop the unneccessary date columns and rename the column
john_hopkins_covid_cases = john_hopkins_covid_cases.drop(john_hopkins_covid_cases.iloc[:, 10:47], axis=1)
john_hopkins_covid_cases.head()
```
